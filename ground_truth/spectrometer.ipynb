```python
from __future__ import division
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (16.0, 16.0)

import numpy as np
import cv2
```

```python
import oceanoptics
spec = oceanoptics.get_a_random_spectrometer()
```

```python
spec.integration_time(0.3)
measurements = []
for i in range(10):
    print i
    s = spec.spectrum()
    measurements.append(s[1])
```

```python
x = s[0]
m = np.mean(measurements, axis=0)
```

```python
plt.plot(x, m)
```

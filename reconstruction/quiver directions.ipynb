```python
with open(os.path.join(BASE_PATH, 'Xs.pkl'), 'rb') as f:
    Xs = cPickle.load(f)
with open(os.path.join(BASE_PATH, 'Ys.pkl'), 'rb') as f:
    Ys = cPickle.load(f)
with open(os.path.join(BASE_PATH, 'Zs.pkl'), 'rb') as f:
    Zs = cPickle.load(f)
with open(os.path.join(BASE_PATH, 'PHIs.pkl'), 'rb') as f:
    PHIs = cPickle.load(f)
with open(os.path.join(BASE_PATH, 'PSIs.pkl'), 'rb') as f:
    PSIs = cPickle.load(f)
with open(os.path.join(BASE_PATH, 'Datas.pkl'), 'rb') as f:
    datas = cPickle.load(f)
```

```python
def quiver(Xs, Ys, Zs, PSIs, PHIs, length = 6000, skip=5):
    p1 = figure(x_range=(-3000, 3000), y_range=(-3000, 3000))
    p2 = figure(x_range=(-3000, 3000), y_range=(-500, 5500))
    cam_ids = Xs.keys()
    for cam_id in cam_ids:
        X, Y, Z, PSI, PHI = \
            Xs[cam_id], Ys[cam_id], Zs[cam_id], PSIs[cam_id], PHIs[cam_id]

        x0 = X[::skip, ::skip].flatten()
        y0 = Y[::skip, ::skip].flatten()
        z0 = Z[::skip, ::skip].flatten()
        psi = PSI[::skip, ::skip].flatten()
        phi = PHI[::skip, ::skip].flatten()

        x1 = x0 + length * np.sin(phi)
        y1 = y0 + length * np.cos(phi)
        z1 = z0 + length * np.cos(psi)
        
        p1.segment(x0, y0, x1, y1)
        p2.segment(x0, z0, x1, z1)
        
        mytext = Label(x=x0[0], y=y0[0], text=cam_id)
        p1.add_layout(mytext)

        mytext = Label(x=x0[0], y=z0[0], text=cam_id)
        p2.add_layout(mytext)

    show(row(p1, p2))
```

```python
def quiver2(datas, length = 6000, skip=5):
    p1 = figure(x_range=(-3000, 3000), y_range=(-3000, 3000))
    p2 = figure(x_range=(-3000, 3000), y_range=(-500, 5500))
    cam_ids = datas.keys()
    for cam_id, data in datas.items():
        x0, y0, z0, phi, psi = \
            [data[i] for i in ('x', 'y', 'z', 'bounding_phi', 'bounding_psi')]

        x1 = x0 + length * np.sin(phi)
        y1 = y0 + length * np.cos(phi)
        z1 = z0 + length * np.cos(psi)
        
        p1.segment(x0, y0, x1, y1)
        p2.segment(x0, z0, x1, z1)
        
        mytext = Label(x=x0, y=y0, text=cam_id)
        p1.add_layout(mytext)

        mytext = Label(x=x0, y=z0, text=cam_id)
        p2.add_layout(mytext)

    show(row(p1, p2))
```

```python
quiver2(datas)
```

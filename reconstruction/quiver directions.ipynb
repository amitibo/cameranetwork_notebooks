```python
from bokeh.io import push_notebook, show, output_notebook
from bokeh.layouts import row
from bokeh.models import Label
from bokeh.plotting import figure
import cPickle
import numpy as np
import os
```

```python
output_notebook()
```

```python
BASE_PATH = r'../../scripts_client/reconstruction/2016_10_23_11_10_00/'
```

```python
with open(os.path.join(BASE_PATH, 'Xs.pkl'), 'rb') as f:
    Xs = cPickle.load(f)
with open(os.path.join(BASE_PATH, 'Ys.pkl'), 'rb') as f:
    Ys = cPickle.load(f)
with open(os.path.join(BASE_PATH, 'PHIs.pkl'), 'rb') as f:
    PHIs = cPickle.load(f)
```

```python
def quiver(Xs, Ys, PSIs, length = 3000, skip=10):
    p = figure(x_range=(-3000, 3000), y_range=(-3000, 3000))
    cam_ids = Xs.keys()
    for cam_id in cam_ids:
        X, Y, PHI = Xs[cam_id], Ys[cam_id], PHIs[cam_id]
        x0 = X[::skip, ::skip].flatten()
        y0 = Y[::skip, ::skip].flatten()
        angle = PHI[::skip, ::skip].flatten()

        x1 = x0 + length * np.sin(angle)
        y1 = y0 + length * np.cos(angle)

        p.segment(x0, y0, x1, y1)

        mytext = Label(x=x0[0], y=y0[0], text=cam_id)
        p.add_layout(mytext)

    show(p)
```

```python
quiver(Xs, Ys, PHIs)
```
